```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

import skopt
from skopt import gp_minimize
```

```python
# Dummy data
x = np.linspace(0, 2, 100)
np.random.seed(0)
scales = np.abs(np.random.randn(10))
noise = np.random.randn(100)

# This param is fixed
A = 10.0

# These are the params 
alpha = 0.2
beta = 1.5
gamma = 0.9

ys = [A*np.exp(-beta*s*x + alpha*noise + gamma) for s in scales]

# The objective is the relative std
L = np.std([y[-1] for y in ys])/np.average([y[-1] for y in ys])
print(L)
for y in ys:
    plt.plot(y)
```

```python
# Optimize with skopt: https://scikit-optimize.github.io/notebooks/hyperparameter-optimization.html

"""
We define the function, we want to minimize
"""
def minimize_this(alpha, beta, gamma):
    ys = [A*np.exp(-beta*s*x + alpha*noise + gamma) for s in scales]
    return np.std([y[-1] for y in ys])/np.average([y[-1] for y in ys])


from skopt.space import Real, Integer
from skopt.utils import use_named_args
"""
We define a parameter space:
parameter type, bounds, distribution and name
"""
space = [
  Real(0.1, 2.0, "log_uniform", name='alpha'),
  Real(0, 2, "uniform", name="beta"),
  Real(0, 10, "uniform", name="gamma")
]

@use_named_args(space)
def objective(**params):
    return minimize_this(**params)

res_gp = gp_minimize(objective, space, n_calls=50, random_state=0, verbose=True)
```
